#Tratamento dos dados

```python
import glob
import pysrt
import re
import os
import pandas as pd

srt_folder = '/content/content/Legendas'
output_folder = '/content/LegendasTexto'
os.makedirs(output_folder, exist_ok=True)

srt_files = glob.glob(os.path.join(srt_folder, '*.srt'))

for file_path in srt_files:
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        text = ''
        for line in lines:
            if re.search(r'^\d+$', line) is None and re.search(r'^\d{2}:\d{2}:\d{2}', line) is None and line.strip() != '':
                text += ' ' + line.strip()
        text = text.lstrip()

    file_name = os.path.basename(file_path)
    name, ext = os.path.splitext(file_name)
    new_file_name = f"{name}_LegendasTexto"
    new_file_path = os.path.join(output_folder, new_file_name)

    with open(new_file_path, 'w', encoding='utf-8') as f:
        f.write(text)

print(f"Processed {len(srt_files)} salvo e extraido para '{output_folder}'")
```
